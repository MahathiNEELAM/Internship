```python
#importing all the necessory libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver  
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import re
import requests

import warnings
warnings.filterwarnings("ignore")
```

```python
#getting the url using driver:
Path = r'C:\Users\admin\Desktop\chromedriver.exe'
driver=webdriver.Chrome(Path)
url=' https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
#import required libraries
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as Ec
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
#Scraping all the required columns and framing them in form of lists:
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
try:
    r=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    n=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    a=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    d=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    v=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in r:
        rank.append(i.text)
    for i in n:
        name.append(i.text)
    for i in a:
        artist.append(i.text)
    for i in d:
        upload_date.append(i.text)
    for i in v:
        views.append(i.text)
except NoSuchElementException :
    loop=[rank,name,artist,upload_date,views]
    for i in loop:
        i.append('_') 
except TimeoutException as e:
    print('Exception Raised: ',e)
  
   
```

```python
#selecting the required data and checking the length of all the lists framed above:
rank=rank[0:30]
name=name[0:30]
artist=artist[0:30]
upload_date=upload_date[0:30]
views=views[0:30]
print(len(rank),len(name),len(artist),len(upload_date),len(views))
```

```python
#Creating a data frame using the above data:
df=pd.DataFrame({'Rank':rank,'Video_name':name,'Artist':artist,'Upload_date':upload_date,'Views':views})
df
```

```python
#importing all the necessory libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver  
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import re
import requests

import warnings
warnings.filterwarnings("ignore")
```

```python
#getting the url using driver:
Path = r'C:\Users\admin\Desktop\chromedriver.exe'
driver=webdriver.Chrome(Path)
url='https://archive.ics.uci.edu/'
driver.get(url)
```

```python
#clicking view all  the dat sets link:
view=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/span/b/a')
view.click()
```

```python
#scraping the data set names in the a bove url:
Datasetname=[]
Datatype=[]
Task=[]
Attributetype=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
dataset=driver.find_elements_by_xpath('//P[@class="normal"]/b/a')
for i in dataset:
    Datasetname.append(i.text)
len(Datasetname)
```

```python
#scraping the all the urls links in the above  page
url=[]
for i in dataset:
    url.append(i.get_attribute('href'))
len(url)
```

```python
#scraping all the data from the urls scrapped above
for i in url:
    driver.get(i)
    time.sleep(5)
    try:
        typ=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table[2]/tbody/tr[1]/td[2]/p')
        task=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table[2]/tbody/tr[3]/td[2]/p')
        att=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table[2]/tbody/tr[2]/td[2]/p')
        inst=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table[2]/tbody/tr[1]/td[4]/p')
        natt=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table[2]/tbody/tr[2]/td[4]/p')
        year=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table[2]/tbody/tr[2]/td[6]/p')
        Datatype.append(typ.text)
        Task.append(task.text)
        Attributetype.append(att.text)
        No_of_instances.append(inst.text)
        No_of_attribute.append(natt.text)
        Year.append(year.text)
    except NoSuchElementException:
        loop=[Datatype,Task,Attributetype,No_of_instances,No_of_attribute,Year]
        for i in loop:
            i.append('_') 
    
```

```python
print(len(Datasetname),len(Datatype),len(Task),len(Attributetype),len(No_of_instances),len(No_of_attribute),len(Year))#checking the length of all lists
```

```python
y=[i.split('-', 1)[0] for i in Year]#getting the year from the date list
y
```

```python
# creating a dataframe:
df=pd.DataFrame({'Dataset_name':Datasetname,
                'Data_type':Datatype[2:624],
                'Task':Task,
                'Attribute_type':Attributetype,
                'No_of_instances':No_of_instances,
                'No_of_attribute':No_of_attribute,
                'Year':y})
df
```
